 # Importy bibliotek

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
import numpy as np
```

# Wczytanie danych

```python



df = pd.read_csv("dane_z_badan_dzwignicy.csv")
display(df.head(12))


print("Podstawowe informacje o danych:")
display(df.info())

print("\nPodstawowe statystyki dla kolumn numerycznych:")
display(df.describe())

print("\nPrzykładowe wiersze:")
display(df.tail())

print("\nLiczba braków danych w każdej kolumnie:")
display(df.isna().sum())
```

# Sprawdzamy braki danych

```python

df_spalinowy = df[df["Maksymalne, chwilowe zużycie ON [l/h]"].notna()]


df_elektryczny = df[df["Maksymalne, chwilowe zużycie energii elektrycznej [kW]"].notna()]

print(f"Liczba rekordów – napęd spalinowy: {len(df_spalinowy)}")
print(f"Liczba rekordów – napęd elektryczny: {len(df_elektryczny)}")
```

```python
# dzielimy dane na spalinowe i elektryczne
df_spalinowy = df[df["Maksymalne, chwilowe zużycie ON [l/h]"].notna()]
df_elektryczny = df[df["Maksymalne, chwilowe zużycie energii elektrycznej [kW]"].notna()]
print(f"Liczba rekordów – napęd spalinowy: {len(df_spalinowy)}")
print(f"Liczba rekordów – napęd elektryczny: {len(df_elektryczny)}")

df_spalinowy = df_spalinowy.drop(df_spalinowy.columns[9:12], axis = 1)
display(df_spalinowy.head())
df_elektryczny = df_elektryczny.drop(df_elektryczny.columns[6:9], axis = 1)
display(df_elektryczny.head())
```

# Usunięcie niepotrzebnych kolumn i uzupełnienie braków danych
Usuwamy kolumne z prędkością wiatru w jednostkach km/h, ponieważ mamy obok kolumnę w jednostkach m/s 



```python
df_elektryczny = df_elektryczny.drop(df_elektryczny.columns[10], axis = 1)
df_spalinowy = df_spalinowy.drop(df_spalinowy.columns[10], axis = 1)
```

```python

# Uzpelniamy braki z lokalnej sredniej 
df_spalinowy_filled = df_spalinowy.copy()

df_spalinowy_filled.iloc[0:3, 8] = df_spalinowy_filled.iloc[3, 8]
lowBound = float(df_spalinowy_filled.iloc[3, 8])
upperBound = float(df_spalinowy_filled.iloc[3, 8])
nanRows = []
for index, row in df_spalinowy_filled.iterrows():
  if row[8] != row[8]:
    nanRows.append(index)
  else:
    if not nanRows:
      lowBound = float(row[8])
    else:
      upperBound = float(row[8])
      for i in nanRows:
        df_spalinowy_filled.iloc[i, 8] = (lowBound + upperBound)/2
      nanRows = []


fig, ax = plt.subplots(1, 2, figsize=(16, 6))


ax[0].scatter(df_spalinowy.Data, df_spalinowy["Cena hurtowa ON 1000l [PLN]"], 
              s=10, color='darkblue', label='Obserwacje')
ax[0].plot(df_spalinowy.Data, df_spalinowy["Cena hurtowa ON 1000l [PLN]"], 
           color='lightblue', linewidth=1, label='Trend')
loc = plticker.FixedLocator([0, len(df_spalinowy)//9, 2*len(df_spalinowy)//9, len(df_spalinowy)//3 - 1])
ax[0].xaxis.set_major_locator(loc)
ax[0].grid(True, linestyle='--', alpha=0.5)
ax[0].set_xlabel("Data [mm/dd/yyyy]")
ax[0].set_ylabel("Cena hurtowa ON 1000l [PLN]")
ax[0].set_title("Ceny ON z brakami")
ax[0].legend()


ax[1].plot(df_spalinowy.Data, df_spalinowy_filled["Cena hurtowa ON 1000l [PLN]"], 
           color='green', linewidth=2, label='Uzupełnione')
ax[1].plot(df_spalinowy.Data, df_spalinowy["Cena hurtowa ON 1000l [PLN]"], 
           color='gray', linewidth=1, label='Oryginalne')
ax[1].scatter(df_spalinowy.Data, df_spalinowy["Cena hurtowa ON 1000l [PLN]"], 
              s=10, color='black', label='Obserwacje')
ax[1].xaxis.set_major_locator(loc)
ax[1].grid(True, linestyle='--', alpha=0.5)
ax[1].set_xlabel("Data [mm/dd/yyyy]")
ax[1].set_ylabel("Cena hurtowa ON 1000l [PLN]")
ax[1].set_title("Ceny ON – uzupełnione średnią lokalną")
ax[1].legend()

fig.suptitle("Porównanie braków i danych uzupełnionych", fontsize=16, fontweight='bold')
fig.tight_layout()
plt.subplots_adjust(top=0.88)
plt.show()



```

 Napęd spalinowy (po lewej)
Silne dodatnie korelacje:

Koszt dzienny silnie koreluje z:

Dzienne zużycie ON [l/8h] (0.77)
Maksymalne chwilowe zużycie ON [l/h] (0.72)
Cena hurtowa ON [PLN] (0.66)
Dzienne zużycie ON [l/8h] silnie koreluje z Maksymalne chwilowe zużycie ON [l/h] (0.94) – logiczne powiązanie.

Istotne zależności operacyjne:

Ciężar ładunku ujemnie koreluje z Długość wysięgnika (-0.60) i Wysokość podnoszenia (-0.54) – może sugerować, że większe ciężary są podnoszone przy krótszym wysięgniku i mniejszej wysokości.

Długość wysięgnika mocno skorelowana z Cena ON (0.89) – możliwa zależność pośrednia przez intensywność pracy.

Słabe lub brak korelacji:

Temperatura, Prędkość wiatru, Ciśnienie – raczej słaba korelacja z kosztami i zużyciem, choć Temperatura koreluje ujemnie z Koszt dzienny (-0.28).

⚡ Napęd elektryczny (po prawej)
Silne dodatnie korelacje:

Koszt dzienny, Dzienne zużycie energii, Maksymalne zużycie energii – wszystkie między sobą bardzo silnie skorelowane (0.85–1). Oznacza to, że koszt zależy głównie od zużycia energii, a nie od innych warunków.

Brak wpływu warunków operacyjnych:

Parametry maszyny (Ciężar ładunku, Długość wysięgnika, Wysokość podnoszenia) mają bardzo słabą korelację z zużyciem i kosztem. Przykładowo: korelacja Ciężar ładunku z Kosztem dziennym to zaledwie 0.033.

Warunki atmosferyczne także nie wpływają znacząco na zmienne docelowe.


W przypadku napędu spalinowego, koszty i zużycie paliwa są mocno uzależnione od warunków pracy maszyny oraz ceny ON, natomiast w napędzie elektrycznym zużycie i koszty są dużo bardziej stabilne i zdeterminowane niemal wyłącznie przez ilość zużytej energii.

1. Średnia efektywność:
Napęd elektryczny wykazuje lepszą średnią efektywność energetyczną niż spalinowy – zużywa mniej energii (kWh) na tonę ładunku niż napęd spalinowy (litrów ON).

2. Rozrzut wartości (wariancja):
Napęd spalinowy ma dużo większy rozrzut – od wartości bardzo niskich aż do ok. 50 l/tonę.

Elektryczny napęd ma węższy zakres wartości, co świadczy o większej stabilności zużycia energii.

3. Obecność odstających wartości (outliers):
W przypadku napędu elektrycznego występuje jedna wartość odstająca (~80 kWh/tonę), co może wskazywać na nietypową sytuację (np. bardzo mały ładunek lub duże zużycie energii z innych przyczyn).

Spalinowy napęd nie wykazuje wyraźnych wartości odstających, ale ma dużą zmienność ogólną.

4. Dolna mediana (IQR):
Większość obserwacji dla elektrycznych napędów znajduje się poniżej wartości typowych dla spalinowych, co wzmacnia wniosek o ich lepszej efektywności.

Napęd elektryczny może być bardziej przewidywalny i kosztowo efektywny, szczególnie w zadaniach o powtarzalnym charakterze.

Napęd spalinowy wykazuje znacznie większe wahania efektywności – może lepiej sprawdzać się w zmiennych warunkach, ale też częściej generuje straty energetyczne.

```python


#  Średnia i odchylenie standardowe


mean_spalinowy_cost = df_spalinowy_filled["Koszt dzienny [PLN]"].mean()
mean_elektryczny_cost = df_elektryczny["Koszt dzienny [PLN]"].mean()

mean_spalinowy_consumption = df_spalinowy_filled["Dzienne zużycie ON [l/8h]"].mean()
mean_elektryczny_consumption = df_elektryczny["Dzienne zużycie energii elektrycznej [kW/8h]"].mean()

std_spalinowy_cost = df_spalinowy_filled["Koszt dzienny [PLN]"].std()
std_elektryczny_cost = df_elektryczny["Koszt dzienny [PLN]"].std()

std_spalinowy_consumption = df_spalinowy_filled["Dzienne zużycie ON [l/8h]"].std()
std_elektryczny_consumption = df_elektryczny["Dzienne zużycie energii elektrycznej [kW/8h]"].std()

print(f"Średnia koszt dzienny - Spalinowy: {mean_spalinowy_cost:.2f}, Elektryczny: {mean_elektryczny_cost:.2f}")
print(f"Średnia zużycie dzienne - Spalinowy: {mean_spalinowy_consumption:.2f}, Elektryczny: {mean_elektryczny_consumption:.2f}")
print("\n")
print(f"Odchylenie standardowe koszt dzienny - Spalinowy: {std_spalinowy_cost:.2f}, Elektryczny: {std_elektryczny_cost:.2f}")
print(f"Odchylenie standardowe zużycie dzienne - Spalinowy: {std_spalinowy_consumption:.2f}, Elektryczny: {std_elektryczny_consumption:.2f}")
```

Napęd spalinowy generuje wyższe koszty dzienne, co jest konsekwencją wyższego kosztu paliwa oraz jego większego zużycia w porównaniu do energii elektrycznej.

Napęd elektryczny wykazuje znacznie mniejsze zmiany w kosztach dziennych, co wskazuje na większą stabilność operacyjną. Jednakże jego zużycie energii elektrycznej jest wyższe, co może wynikać z konieczności ładowania akumulatorów lub różnic w efektywności.

Warto zauważyć, że napęd elektryczny jest bardziej stabilny pod względem zarówno kosztów, jak i zużycia energii, co może sugerować większą przewidywalność i efektywność w długoterminowej eksploatacji.

Napęd spalinowy ma większą zmienność kosztów, co może wymagać lepszego zarządzania w kontekście fluktuacji cen paliwa oraz zmieniających się warunków operacyjnych.

```python
# przeliczenie zużycia energii na MJ
df_spalinowy_filled["Zużycie energii [MJ]"] = df_spalinowy_filled["Dzienne zużycie ON [l/8h]"] * 35.8
df_elektryczny["Zużycie energii [MJ]"] = df_elektryczny["Dzienne zużycie energii elektrycznej [kW/8h]"] * 3.6

# przeliczenie efektywności na MJ/tonę
df_spalinowy_filled["Efektywność [MJ/tonę]"] = df_spalinowy_filled["Zużycie energii [MJ]"] / df_spalinowy_filled["Ciężar ładunku [T]"]
df_elektryczny["Efektywność [MJ/tonę]"] = df_elektryczny["Zużycie energii [MJ]"] / df_elektryczny["Ciężar ładunku [T]"]


mean_spalinowy_cost = df_spalinowy_filled["Koszt dzienny [PLN]"].mean()
mean_elektryczny_cost = df_elektryczny["Koszt dzienny [PLN]"].mean()

mean_spalinowy_energy = df_spalinowy_filled["Zużycie energii [MJ]"].mean()
mean_elektryczny_energy = df_elektryczny["Zużycie energii [MJ]"].mean()

mean_spalinowy_eff = df_spalinowy_filled["Efektywność [MJ/tonę]"].mean()
mean_elektryczny_eff = df_elektryczny["Efektywność [MJ/tonę]"].mean()


import matplotlib.pyplot as plt
import seaborn as sns

labels = ['Koszt dzienny [PLN]', 'Zużycie energii [MJ]', 'Efektywność [MJ/tonę]']
spalinowy_values = [mean_spalinowy_cost, mean_spalinowy_energy, mean_spalinowy_eff]
elektryczny_values = [mean_elektryczny_cost, mean_elektryczny_energy, mean_elektryczny_eff]

x = range(len(labels))
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
plt.bar([i - 0.2 for i in x], spalinowy_values, width=0.4, label='Spalinowy', color='steelblue')
plt.bar([i + 0.2 for i in x], elektryczny_values, width=0.4, label='Elektryczny', color='darkorange')

plt.xticks(x, labels)
plt.ylabel("Średnia wartość")
plt.title("Porównanie napędów – koszt, zużycie energii, efektywność (MJ)")
plt.legend()
plt.tight_layout()
plt.show()
```
