 # Monte Carlo

```python
import numpy as np
import random
import matplotlib.pyplot as plt

```

```python


def MonteCarlo(N):
    # tworzymy tablice w których będą zapisywane wartosci do pozniejszej wizualizacji 
    x_in =[]
    x_out = []
    y_in = []
    y_out =[]
    
    res = 0
    for i in range(N):
        x = random.uniform(0,1)
        y = random.uniform(0,1)
        odl = np.sqrt(x**2 + y**2)
        if odl <=1:
            res +=1
            x_in.append(x)
            y_in.append(y)
        else:
            x_out.append(x)
            y_out.append(y)
    pole = res/ N
    pi = pole * 4
    print(f"Przybliżona wartość pi wynosi: {pi}")
    return pi, x_in, y_in, x_out, y_out



MonteCarlo(1000)

    
```
